```python
user_list = pd.read_csv("put_csv/user.csv")
for idx in range(user_list.shape[0]):
    username      = user_list['username'][idx]
    user_id       = user_list['user_id'][idx]
    password      = user_list['password'][idx]
    mobile_number = user_list['mobile_number'][idx]
    email         = user_list["email"][idx]
    point         = user_list['point'][idx]
    hashed_password = bcrypt.hashpw(
                    password.encode('utf-8'),
                    bcrypt.gensalt()
    ).decode()
    User.objects.create(
        username      = username,
        user_id       = user_id,
        password      = hashed_password,
        mobile_number = mobile_number,
        email         = email,
        point         = point
)
```
